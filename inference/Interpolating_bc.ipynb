Interpolation of timepoints in bc01 and bc02 that are present in gln01-gln09

```python
import pandas as pd
bc = pd.read_csv("../PopPropForABC_bc.csv")
gln = pd.read_csv("../PopPropForABC.csv")
bc_gen = pd.read_csv("../generations_bc.csv")
gln_gen = pd.read_csv("../generations.csv")
```

```python
bc = pd.concat([bc, bc_gen], axis=1)
bc
```

```python
gln=pd.concat([gln, gln_gen], axis=1)
gln
```

```python
bc_new = pd.concat([bc,pd.merge(left=gln_gen, right=bc, how="left", left_on='Generation', right_on='Generation')], axis=0)
```

```python
bc_new = bc_new.drop_duplicates().sort_values(by="Generation")
bc_new = bc_new.set_index('Generation')
bc_new
```

```python
bc_new = bc_new.interpolate('values')
bc_new
```

```python
import matplotlib.pyplot as plt
plt.plot(bc['Generation'],bc['bc01'], color='k')
```

```python
plt.plot(bc_new.index,bc_new['bc01'], color='red')
```

```python
import matplotlib.pyplot as plt
plt.plot(bc['Generation'],bc['bc02'], color='k')
```

```python
plt.plot(bc_new.index,bc_new['bc02'], color='red')
```

```python
#bc_new['Generation'] = bc_new.index
all_exp = pd.merge(left=gln, right=bc_new, how="left", left_on='Generation', right_on='Generation')
```
