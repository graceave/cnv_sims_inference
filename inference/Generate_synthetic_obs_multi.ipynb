```python
import numpy as np
from cnv_simulation import CNVsimulator_multiChemo, CNVsimulator_multiWF
import matplotlib.pyplot as plt
```

```python
generation=np.genfromtxt("../generations.csv",delimiter=',', skip_header=1,dtype="int64")
N = 3.3e8
s_snv=0.001 
m_snv=1e-5 
reps=11
```

```python
cnv_params = np.array([10,np.log10(0.01),np.log10(1e-7)])
states, s_cnv = CNVsimulator_multiChemo(reps, s_snv, m_snv, generation, parameters=cnv_params, return_s=True, seed=None)
#CNVsimulator_multiWF(reps, N, s_snv, m_snv, generation, parameters=cnv_params, return_s=True, seed=None)
```

```python
plt.plot(generation,states.T,linewidth=1, color = 'black', alpha=0.6)
```

```python
all_data = np.append(states, np.log10(s_cnv).reshape(reps, 1), axis=1)
all_data = np.append(all_data, np.tile(np.log10(1e-7),(reps,1)), axis=1)
np.savetxt("Chemo_shape10_scale0.01_mut7_observations_1.csv", all_data, delimiter=',')
```
